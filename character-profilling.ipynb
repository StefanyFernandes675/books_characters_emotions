```python
import spacy
import nltk
import PyPDF2
import time
from nltk.sentiment import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
```

```python
start_time = time.time()
```

```python
# Open the PDF file in read-binary mode
pdf_file = open('/content/death_on_the_Nile.pdf', 'rb')

# Create a PDF reader object
pdf_reader = PyPDF2.PdfReader(pdf_file)

# Extract the text from each page
text = ''
for page in pdf_reader.pages:
    text += page.extract_text()

# Close the PDF file
pdf_file.close()

# Print the extracted text
print(text)
```

Pre-trained english model

```python
nlp = spacy.load('en_core_web_sm')
```

## Hercule Poirot

```python
# Identify sentences spoken by Hercule Poirot
hercule_poirot_sentences = []
for sentence in nlp(text).sents:
    if 'Hercule Poirot' in sentence.text or 'Poirot' in sentence.text:
        hercule_poirot_sentences.append(sentence.text)
```

```python
print(hercule_poirot_sentences)
```

```python
# Perform emotional character profiling analysis on Hercule Poirot's dialogue
sia = SentimentIntensityAnalyzer()
hercule_poirot_sentiment = 0
for sentence in hercule_poirot_sentences:
    print(sentence)
    scores = sia.polarity_scores(sentence)
    print('Sentiment score:', scores)
    hercule_poirot_sentiment += scores['compound']
```

```python
if hercule_poirot_sentiment > 0:
  sentiment = "positive"
elif hercule_poirot_sentiment < 0:
  sentiment = "negative"
else:
  sentiment = "neutral"
print(hercule_poirot_sentiment)
print("Hercule Poirot's overall sentiment:", sentiment)
```
